# funciones



```python
tDocente=ndf2['Docente'].unique()
tAlumno=ndf2['Alumno'].unique()
```

```python
# iniciamos un diccionario, index con codigo docente y values 0 
# que se modificará conforme se agreguen alumnos a la lista del docente
nro=np.zeros(41)
cantAlumnDocente=dict(zip(tDocente,nro))
cantAlumnDocente
```

```python
#inicializacion de listas que usaremos:
#lista estudiantes
listEstudianteA=[]
#diccionario donde se iran agregando el alumno y docente para la lista de la infolimpieza
listAlumnoDocente={}
```

```python
#funcion para saber si el codigo del alumno que estamos viendo
#ya a sido asignado con un docente

def tienedocente(alumno):
  if alumno in listEstudianteA:
    return True
  else:
    return False
```

```python
##funcion para asignarle docente a cada alumno

#teniendo la cantidad de estudiantes y de alumnos, 
#calculamos que debe asignarse a lo mas 15 alumnos a cada docente

def asignardocente(a,d):
  if (tienedocente(a)==True):
    return listAlumnoDocente[a] 
  elif (cantAlumnDocente[d] < 15):
      cant=cantAlumnDocente[d]
      listEstudianteA.append(a)
      cantAlumnDocente[d] = cant + 1
      listAlumnoDocente[a] = d
      return d
  else:
      return np.NaN

```

```python
#se crea una columna para cada alumno en la que
#se tenga el codigo del docente al que irá pertnecer a su lista
 
ndf2.apply(lambda x: asignardocente(x[0],x[4]), axis=1,  result_type='expand')
```
